# Exercise 1

```python
# Some data for plotting
from matplotlib import pyplot as plt
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Exercise 2

```python
import numpy as np
def generate_x_y(resolution):
    increment=50/resolution
    hor=np.array([increment*number for number in range(resolution+1)])
    ver=np.array([np.exp(-1*number/10)*np.sin(number) for number in hor])
    return np.array([hor,ver])

res=1000 #number of (x,y) values calculated (not including (0,0))
x=generate_x_y(res)[0]
y=generate_x_y(res)[1]
plt.plot(x, y, color='blue')
plt.xlabel('x', color='darkorange',fontsize=15)
plt.ylabel('y', rotation='horizontal',color='darkorange',fontsize=15)
plt.title('$y=e^{-x/10}sin(x)$',fontsize=20)
plt.grid(alpha=.25)
plt.ylim(-1,1)
```
